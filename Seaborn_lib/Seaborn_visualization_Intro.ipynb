```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set_style("whitegrid")
```

```python
df=pd.read_csv("market_fact.csv")
```

```python
#simple density plot
sns.distplot(df['Shipping_Cost'])
plt.show()
```

```python
sns.distplot(df['Shipping_Cost'][:200],rug=True)
plt.show()
```

```python
sns.distplot(df['Shipping_Cost'][:200],rug=True,hist=False)
plt.show()
```

```python
plt.figure(1)
plt.subplot(2,2,1)
sns.distplot(df['Sales'])
plt.title("Sales")
#plt.show()

plt.subplot(2,2,2)
sns.distplot(df['Profit'])
plt.title("Profit")
#plt.show()

plt.figure(2)
plt.subplot(2,2,1)
sns.distplot(df['Order_Quantity'])
plt.title("Order_Quantity")
#plt.show()

plt.subplot(2,2,2)
sns.distplot(df['Shipping_Cost'])
plt.title("Shipping_Cost")
#plt.show()
```

```python
sns.boxplot(df['Order_Quantity'])
plt.title("Order_Quantity")
plt.show()
```

```python
sns.boxplot(y=df['Order_Quantity'])
plt.title("Order_Quantity")
plt.show()
```

# Bivariate Analysis

```python
sns.jointplot('Sales','Profit',df)
plt.show()
```

```python
#very skewed
#can be better visualized by removing outliers and for that we can put some cndition
df=df[(df.Profit<10000)&(df.Sales<20000)]
sns.jointplot('Sales','Profit',df)
plt.show()
```
